```python
# Build the Model
# encoder_inputs = 
normalization_layer = tf.keras.layers.Lambda(lambda x: (x - tf.reduce_min(x)) / (tf.reduce_max(x) - tf.reduce_min(x)))
z_score = tf.keras.layers.Lambda(lambda x: (x - tf.reduce_mean(x)) / tf.math.reduce_std(x))

model = tf.keras.models.Sequential([
        # tf.keras.layers.Embedding(input_dim=4, output_dim=4, input_length=4),
        # tf.keras.layers.LSTM(units=64),
        z_score,
        normalization_layer,
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dense(512, activation='relu'),
        # Menambahkan layer output dengan jumlah kelas yang sesuai
        tf.keras.layers.Dense(18, activation='softmax'),
        # tf.keras.layers.Dropout(0.2))  
    ], name="Classification"
    )
#  # Print the model summary 
model.compile(optimizer=tf.keras.optimizers.Adam(0.01),
              loss='categorical_crossentropy',
              metrics=['accuracy']
              )
```
