```python
import numpy as np
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("Wheater-Encode-22.csv")

x = data.iloc[:, :4 ].values
y = data.iloc[:,4:].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
print("Data shape:", x_train.shape)
print("Labels shape:", y_train.shape)
```

```python
# Build the Model
# encoder_inputs = 
normalization_layer = tf.keras.layers.Lambda(lambda x: (x - tf.reduce_min(x)) / (tf.reduce_max(x) - tf.reduce_min(x)))
z_score = tf.keras.layers.Lambda(lambda x: (x - tf.reduce_mean(x)) / tf.math.reduce_std(x))

model = tf.keras.models.Sequential([
        # tf.keras.layers.Embedding(input_dim=4, output_dim=4, input_length=4),
        # tf.keras.layers.LSTM(units=64),
        z_score,
        normalization_layer,
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dense(512, activation='relu'),
        # Menambahkan layer output dengan jumlah kelas yang sesuai
        tf.keras.layers.Dense(18, activation='softmax'),
        # tf.keras.layers.Dropout(0.2))  
    ], name="Classification"
    )
#  # Print the model summary 
model.compile(optimizer=tf.keras.optimizers.Adam(0.01),
              loss='categorical_crossentropy',
              metrics=['accuracy']
              )
```

```python
# Melakukan pelatihan model
model.fit(x_train, y_train, epochs=150, batch_size=72, validation_data=(x_test, y_test))
```

```python
model.summary()
```

```python
# Melakukan prediksi dengan model
data = input_data = np.array([[24.4, 84, 5, 90], [22.0, 80.1, 4.8, 270.0], [25.3, 70.5, 5.5, 210.0]])
predictions = model.predict(data)

# Melihat hasil prediksi
predicted_labels = np.argmax(predictions, axis=1)
print(predicted_labels)
```
